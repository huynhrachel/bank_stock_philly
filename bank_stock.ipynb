```python
list_banks = ['JPMC', 'BOA', 'PNC']
```

```python
bank_stocks = pd.concat([JPMC, BoA, PNC], axis=1, keys = list_banks)
bank_stocks.head()
```

```python
bank_stocks.columns = ['_'.join(col) for col in bank_stocks.columns]
print(bank_stocks.head())
```

```python
bank_stocks.filter(like='Close', axis=1).max()
```

```python
value_banks = pd.DataFrame()
```

```python
print(bank_stocks.columns)  # Kim tra danh sch ct
print(bank_stocks.columns.nlevels)  # Kim tra s cp ca MultiIndex
```

```python
def rename_columns(col):
    parts = col.split("_")
    if len(parts) == 3:
        return f"{parts[1]}_{parts[0]}"
    return col
bank_stocks.columns = [
    rename_columns(col).replace("_JPM", "").replace("_BAC", "").replace("_PNC", "")
    for col in bank_stocks.columns
]
print(bank_stocks.columns)



```

```python
for name in list_banks:
    value_banks[name + '_value_bank'] = bank_stocks[f'{name}_Close'].pct_change()
value_banks.head()
```

```python
import seaborn as sns
sns.pairplot(value_banks)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(30,20))

sns.displot(value_banks.loc['2023-01-01':'2024-12-31']['BOA_value_bank'],
            color='green', bins=100)
```

```python
close_columns = [col for col in bank_stocks.columns if "Close" in col]
bank_stocks[close_columns].plot(figsize=(12,8))
```

```python
plt.figure(figsize=(12,6))
ax = JPMC['Close'].loc['2023-01-01':'2024-12-31'].rolling(window=30).mean().plot(label='MA30',color='red')
df = JPMC['Close'].loc['2023-01-01':'2024-12-31'].plot(label='JPMC CLOSE', color='brown')

plt.legend()
plt.show()
```

```python
close_columns = [col for col in bank_stocks.columns if 'Close' in col]
sns.heatmap(bank_stocks[close_columns].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap for Close Prices")
plt.show()
```

```python
close_columns = [col for col in bank_stocks.columns if 'Close' in col]
sns.clustermap(bank_stocks[close_columns].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Clustermap for Close Prices")
plt.show()
```
